```python
covariance_Y = model[0][0][-1]
matrices = covariance_Y[0]
```

```python
matrices[1]
```

```python
def corr_matrix(covar):
    rows, cols = covar.shape
 
    corr_mat = np.zeros((cols, cols))
    
    for i in range(cols):
        for j in range(cols):
            corr_mat[i][j] = covar[i][j] / ((np.sqrt(covar[i][i])) * (np.sqrt(covar[j][j])))
    mask = np.triu(np.ones_like(corr_mat, dtype=bool))
    corr_mat[mask]=np.nan
    return corr_mat
```

```python
corr_matrices = [corr_matrix(matrix) for matrix in matrices]
```

```python
fig = px.scatter_mapbox(butterfly_covar,
        lon = butterfly_covar['Longitude'],
        lat = butterfly_covar['Latitude'], zoom=10, hover_data=["Building", "Urban Vegetation"], width = 900, height = 900)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
#fig.show() 
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
matrix_list = corr_matrices
fig_list = [px.imshow(matrix, color_continuous_scale='RdBu_r', zmin=-1, zmax=1, text_auto=".2f", aspect="auto",
                      x=but_names_x, y=but_names_y) for matrix in matrix_list]

modal = html.Div(
    [
        dbc.Modal(
            [
                dbc.ModalHeader("Modal Header"),
                dbc.ModalBody([
                    html.H4(id='hover_info'),
                    dcc.Graph(
                        id='modal_graph',
                        style={'width': '100%', 'height': '500px'}
                    )
                ]),
                dbc.ModalFooter(
                    dbc.Button("Close", id="close", className="ml-auto")
                ),
            ],
            id="modal",
            centered=True,
            size="xl",
            style={"width": "100%", "height": "100vh", "margin": 0} 
        ),
    ]
)

app.layout = dbc.Container([
    dbc.Row([
        html.H2('Correlation Map')
    ]),
    dbc.Row([
        dbc.Col(
            dcc.Graph(
                id='my_graph',
                figure=fig
            )
        )
    ],justify='center'),
    modal
])

@app.callback(
    Output("modal", "is_open"),
    Output("hover_info","children"),
    Output("modal_graph", "figure"),
    [Input("my_graph", "clickData"), Input("close", "n_clicks")],
    [State("modal", "is_open")],
)
def toggle_modal(hover_data,close_button, is_open):
    if hover_data or close_button:
        point_index =  hover_data['points'][0]['pointIndex']
        print(point_index)
        lon = hover_data['points'][0]['lon']
        lat = hover_data['points'][0]['lat']
        #custom_data = hover_data['points'][0]['customdata']
        #building = custom_data[0]
        #vegetation = custom_data[1]
        text = f"lat: {lat:.4f}, lon: {lon:.4f}"
        selected_figure = fig_list[point_index]
        selected_figure.update_layout(
        plot_bgcolor='rgba(0, 0, 0, 0)',  # Transparent plot background
        paper_bgcolor='rgba(0, 0, 0, 0)'  # Transparent paper background
        )
        selected_figure.update_layout( yaxis = dict(tickfont = dict(size=5)), xaxis = dict(tickfont = dict(size=5)))
        selected_figure.update_xaxes(tickvals=np.arange(len(labels)), ticktext=but_names_x, dtick = 1)
        selected_figure.update_yaxes(tickvals=np.arange(len(labels)), ticktext=but_names_y, dtick = 1)
        return not is_open,text, selected_figure
    return is_open,None, px.imshow(np.zeros((5, 5)))

if __name__ == "__main__":
    port = 8054
    app.run_server(debug=True, port=port)
    server_url = f'http://127.0.0.1:{port}'
    webbrowser.open(server_url)
```
